```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support.ui import Select

from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException, ElementNotVisibleException
from selenium.webdriver.support import expected_conditions as EC

from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
from selenium.webdriver.chrome.webdriver import WebDriver as ChromeWebDriver
from selenium.webdriver.chrome.options import Options
import csv
import re
```

```python
def followup_comment(item):
    # Vote: [Column Number] + “ - “ + [Vote and Type] + CRLF + 
    # “Submitted by: ” [Submitted by]  + “(“ + [Organization] + “) “ +
    # “ On behalf of: “ + [On behalf of] + “(“ + [Commenter e-mail] + “)”  + 
    # CRLF + “In person” if [In Person Resolution Requested] is “Yes”) + CRLF + 
    # “Comment: “ + [Ballot Comment].  (Note: Only include labels and line breaks
    # if there’s content for that label/line.)
    ret = """Vote: %s - %s"""%(item["Comment \nNumber"], item["Vote and Type"])
    
    if item["Submitted By"] or item["Organization"]:
        org = item["Organization"]
        if org:
            org = "(%s)"%org
        ret += "\nSubmitted by: %s  %s"%(item["Submitted By"], org)

    if item["On behalf of"]  or item["On Behalf of Email"]:
        email = item["On Behalf of Email"]
        if email:
            email = "(%s)"%email
        ret += "\nOn behalf of %s %s"%(item["On behalf of"], email)
        
    if item["In person resolution requested"] :
        ret += "\nIn person: %s"%(item["In person resolution requested"])
        
    if item["Comments"]:
        ret += "\n---\nComment:\n%s"%(item["Comments"])
        
    return ret
```

```python
def select_box(label):
    label_elt = driver.find_element(By.XPATH, '//strong[text()="'+label+'"]')
    return Select(label_elt.find_element(By.XPATH, '../select'))

def input_box(label):
    label_elt = driver.find_element(By.XPATH, '//strong[text()="'+label+'"]')
    return label_elt.find_element(By.XPATH, '../input')

def fill_iframe(divid, text):
    followup_iframe = WebDriverWait(driver, 5).until(EC.frame_to_be_available_and_switch_to_it((By.CSS_SELECTOR, "#"+divid+" iframe")))
    followup_box = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.CSS_SELECTOR, "body[contenteditable='true']")))
    followup_box.clear()
    followup_box.send_keys(text)
    driver.switch_to_default_content()
    
def fill_followup(text):
    fill_iframe("cke_text_body", text)
        
def fill_details(text):
    fill_iframe("cke_details", text)

def save_changes():
    save_button = driver.find_element(By.PARTIAL_LINK_TEXT, 'Save changes')
    save_button.click()
    WebDriverWait(driver, 10).until(EC.title_contains("Browse Tracker Item"))

def submit_item():
    add_button = driver.find_element(By.PARTIAL_LINK_TEXT, 'Add')
    add_button.click()
    WebDriverWait(driver, 5).until(EC.text_to_be_present_in_element((By.CSS_SELECTOR, "span.feedback"), "Successfully added"))
    feedback = driver.find_element(By.CSS_SELECTOR, "span.feedback")
    return re.findall("(\d+)",feedback.text)[0]

def set_ballot_weight(item):
    ballot_weight = select_box("Ballot-weight")
    current_selection = ballot_weight.all_selected_options[0].text
    desired_selection = item['Vote and Type']
    if desired_selection.startswith("A") and current_selection.startswith("None"):
        ballot_weight.select_by_visible_text("Affirmative")
    elif desired_selection.endswith("Mi") and not current_selection.endswith("Major"):
        ballot_weight.select_by_visible_text("Negative-Minor")
    elif desired_selection.endswith("Mj"):
        ballot_weight.select_by_visible_text("Negative-Major")
            
def set_in_person(item):
    in_person = select_box("In-Person?")
    current_selection = in_person.all_selected_options[0].text
    in_person_desired = item["In person resolution requested"] == "Yes"        
    if in_person_desired and not current_selection.startswith("Yes"):
         in_person.select_by_visible_text("Yes")

def navigate_to(url, form):
    print "Navigating to %s"%url
    driver.get(url)
    element = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.NAME, form)))
    
def navigate_to_edit(item_id):
    navigate_to(edit_url%item_id, "TrackerItemEdit")

    
def navigate_to_add(tracker_id):
    navigate_to(add_url%tracker_id, "TrackerItemAdd")

def update_tracker_item(item):
    try:
        navigate_to_edit(item['Tracker Num.'])
        set_ballot_weight(item)
        set_in_person(item)
        fill_followup(followup_comment(item))
        save_changes()
        print "Done with", item["Comment \nNumber"]

    except (ElementNotVisibleException, TimeoutException) as e:
        print "Can't find tracker %s from comment %s. "%(item['Tracker Num.'], item['Comment \nNumber'])
```

```python
def issue_details(item):
    #Details: “Submitted by: ” [Submitted by]  + “(“ + [Organization] + “) “ + 
    # “ On behalf of: “ + [On behalf of] + “(“ + [Commenter e-mail] + “)”  + CRLF 
    # + “In person” if [In Person Resolution Requested] is “Yes”) + CRLF +
    # “Existing Wording: “ + [Existing Wording][Ballot Comment] +CRLF + 
    # “Proposed Wording: “ + [Proposed Wording] + CRLF + 
    # “Comment: “ + [Ballot Comment].

    ret = ""
    
    if item["Submitted By"] != "" or item["Organization"] != "":
        org = item["Organization"]
        if org != "":
            org = "(%s)"%org
        ret += "\nSubmitted by: %s  %s"%(item["Submitted By"], org)

    if item["On behalf of"] != "" or item["On Behalf of Email"] != "":
        email = item["On Behalf of Email"]
        if email != "":
            email = "(%s)"%email
        ret += "\nOn behalf of %s %s"%(item["On behalf of"], email)
        
    if item["Existing Wording"]:
        ret += "\nExisting Wording: %s"%item["Existing Wording"]

    if item["Proposed Wording"]:
        ret += "\nProposed Wording: %s"%item["Proposed Wording"]
        
    if item["Comments"]:
        ret += "\n---\nComment:\n%s"%(item["Comments"])
        
    return ret

def select_all(select_elt, values):
    for v in values:
        try:
            select_elt.select_by_visible_text(v)
        except:
            print "couldn't fill", v
            pass
    if len(select_elt.all_selected_options) > 1:
        select_elt.deselect_by_visible_text("None")

"""
If the [Tracker #] column is blank, create a new tracker item as follows:
Summary: [Column Number] + “ - “ + [Summary]
Details: “Submitted by: ” [Submitted by]  + “(“ + [Organization] + “) “ + “ On behalf of: “ + [On behalf of] + “(“ + [Commenter e-mail] + “)”  + CRLF + “In person” if [In Person Resolution Requested] is “Yes”) + CRLF + “Existing Wording: “ + [Existing Wording][Ballot Comment] +CRLF + “Proposed Wording: “ + [Proposed Wording] + CRLF + “Comment: “ + [Ballot Comment].  (Note: Only include labels and line breaks if there’s content for that label/line.)
Specification: [Standard]
url: [URL]
Section number: [Section]
Resource(s): [Resource(s)]
HTML Page: [HTML Page name(s)]
Reviewing Work Group(s) [Disposition WG] (separated by “,” if there’s more than one)
Status: “Triaged”
Vote Strength: [Vote and Type] (Neg-Mi, Typo = Negative-Minor; Neg-Mj = Negative-Major, everything else = affirmative)
Category: [Sub-Category], if present.  Otherwise, “Typo” if [Vote and Type]=”A-T”, “Question” if [Vote and Type]=”A-Q” or “Comment” if [Vote and Type]=”A-S”
In Person?: true if [In person resolution requested]=”Yes”
Schedule: [Schedule]
Grouping: [Comment grouping]

First comment: [Triage Note]
"""

def create_tracker_item(item):
    navigate_to_add(tracker_id)
    summary_val = "DSTU2 Comment #%s "%item["Comment \nNumber"]
    if item["Summary"]:
        summary_val += " - %s"%item["Summary"]
    summary_field = driver.find_element(By.NAME, "summary")
    summary_field.send_keys(summary_val)
    
    if item["Page or Section URL"]:
        input_box("url").send_keys(item["Page or Section URL"])
        
    if item["Section"]:
        input_box("Section number").send_keys(item["Section"])
        
    if item["Resource(s)"]:
        rs = item["Resource(s)"]
        rs = re.split('\W+', rs)
        select_all(select_box("Resource(s)"), rs)
        
    if item["HTML Page name(s)"]:
        print "Selectin gall htmla", item["HTML Page name(s)"]
        select_all(select_box("HTML Page(s)"), [item["HTML Page name(s)"]])
        
    fill_details(issue_details(item))
    new_item_id = submit_item()

    navigate_to_edit(new_item_id)
    select_all(select_box("Status"), ["Triaged"])
    set_ballot_weight(item)

    category = item["Sub-category"]
    vote = item["Vote and Type"]
    if not category:
        if vote == "A-T": category = "Typo"
        elif vote == "A-Q": category = "Question"
        elif vote == "A-S": category = "Comment"
    if category:
        select_all(select_box("Category"), [category])

    set_in_person(item)

    if item["Comment grouping"]:
        select_all(select_box("Group"), [item["Comment grouping"]])
        
    if item["Triage Note"]:
        fill_followup("Triaged: %s"%item["Triage Note"])

    save_changes()
```
